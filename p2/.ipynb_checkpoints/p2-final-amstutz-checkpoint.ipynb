## 1) nettoyage des données

### 1.1) estimation des ages Nan

nous remarquons que dans la colonne "age", des valeurs ne sont pas renseigné. il faut donc leurs donner une valeur

ainsi, nous allons estimé les ages Nan par la medianne des ages du meme sexe et meme classe. 
etapes :
* pour simplifier le travail, ajout d'une colonne SexBinary pour avoir des valeurs numerique (1=male|0=female)
* creation d'un tableau des mediannes avec deux lignes SexBinary(1|0) et 3 colonnes Pclass(1|2|3)
* on remplace les valeurs Nan par la medianne correspondante a SexBinary et Pclass

###### etape 1 : colonne SexBinary

```python
dfTrain['SexBinary'] = dfTrain['Sex'].map({'male': 1, 'female': 0})
```

```python
dfTrain.info()
dfTrain.head(2)
```

nous avons bien notre nouvelle colonne avec des valeurs binaire qui correspondent au sexe

###### etape 2 : tableau des mediannes

```python
#init tableau 
tabMediannes = np.zeros((2, 3), dtype='f')

for sexIndex in range(0, 2):
    for classIndex in range(1, 4):
        tabMediannes[sexIndex,classIndex-1] = dfTrain[(dfTrain['SexBinary'] == sexIndex) & (dfTrain['Pclass'] == classIndex)]['Age'].dropna().median()
```

```python
#affichage des valeurs
for classIndex in range(0, 3):
    print("la medianne d'age des hommes de classe ",classIndex," est de ",tabMediannes[1,classIndex])
for classIndex in range(0, 3):
    print("la medianne d'age des femmes de classe ",classIndex," est de ",tabMediannes[0,classIndex])
```

###### etape 3 : on remplace les valeurs Nan

 *remarque : nous sommes obligé de faire une copie de la colonne age pour pouvoir ecrire dedans * 

```python
#copie
dfTrain['AgeCP'] = dfTrain['Age'].copy()

#pour chaque ligne avec age = Nan , on remplace par la valeur correspondante de tabMediannes
for index, age in enumerate(dfTrain.Age):
    if math.isnan(age) :
        dfTrain.loc[index,'AgeCP'] = tabMediannes[dfTrain.SexBinary[index],dfTrain.Pclass[index]-1]
```

```python
#affichage
dfTrain.describe()
```

et voila! il n'y a pas de valeur Nan dans AgeCP :)

#### 2.2) classe du passager

```python
dfTrain['Pclass'].hist()
plb.show()
```

notre dataset contient des passager dans les 3 classes, ce qui est une bonne chose! nous avons beaucoup plus de 3ieme classe mais cela est en adequation avec le pourcentage de 1ier/2nd/3ieme classe disponible.
nous allons maintenant pouvoir voir le pourcentage de survivant pour chacune des classes

```python
labels = 'deces', 'vie'
colors = ['grey', 'lightskyblue']
explodes = [0,0]


for classIndex in range(1, 4):
    nbDeces = dfTrain.loc[((dfTrain.Pclass == classIndex) & (dfTrain.Survived == 0))].size
    nbVie = dfTrain.loc[((dfTrain.Pclass == classIndex) & (dfTrain.Survived == 1))].size
    print("pourcentages deces/vie pour la classe ",classIndex)
    sizes = [nbDeces, nbVie]
    plt.pie(sizes, explode=explodes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
    plt.axis('equal')
    plt.show()
```

il parrait maintenant evident meilleur est la classe, plus la chance de survie est importante...

#### 2.3) age du passager

nous allons ici etudier le nombre de deces par age tranché de 5 en 5 ans.
commencons par le nombre de passager par tranches de 5 ans

```python
prctDeces = []
ages = []

for ageMax in range(5,85,5):
    ages.insert((int((ageMax/5)-1)),ageMax)
    prctDeces.insert((int((ageMax/5)-1)), dfTrain.loc[((dfTrain.AgeCP <= ageMax) & (dfTrain.AgeCP >= (ageMax-5)))].size)


plt.figure()
plt.title('age v nbPassager')
plt.xlabel('Age')
plt.ylabel('Passager')
plt.plot(ages, prctDeces, 'k.')
plt.axis([0, 80, 0, 4000])
plt.grid(True)
plt.show()
```

```python
prctDeces = []
ages = []

for ageMax in range(5,85,5):
    ages.insert((int((ageMax/5)-1)),ageMax)
    prctDeces.insert((int((ageMax/5)-1)), dfTrain.loc[((dfTrain.AgeCP <= ageMax) & (dfTrain.AgeCP >= (ageMax-5)) & (dfTrain.Survived == 0))].size/(dfTrain.loc[((dfTrain.AgeCP <= ageMax) & (dfTrain.AgeCP >= (ageMax-5)))].size))


plt.figure()
plt.title('age v nbPassager')
plt.xlabel('Age')
plt.ylabel('Passager')
plt.plot(ages, prctDeces, 'k.')
plt.axis([0, 80, 0,1])
plt.grid(True)
plt.show()
```
