```python
import numpy as np
import scipy.stats as ss
import matplotlib.pyplot as plt
import sklearn
import pandas as pd
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
%matplotlib inline

model = LinearRegression()
elec=[[69032],[69035],[69044]]
heures=[[13],[16],[21]]
model.fit(heures, elec)

plt.figure()
plt.title('kwh v heure')
plt.xlabel('heure')
plt.ylabel(u'Kwh')
plt.scatter(heures, elec, color='black')
plt.plot(heures, model.predict(heures), linewidth=3)
plt.axis([0, 24, 69020, 69060])
plt.grid(True)

plt.show()

heure = 7
heure2 = 24+7
print(u' à {d}h -> {px:.2f} kwh'.format(
        d=heure,
        px=model.predict([[heure]])[0][0]))
print(u' à {d}h -> {px:.2f} kwh'.format(
        d=heure2,
        px=model.predict([[heure2]])[0][0]))
```
